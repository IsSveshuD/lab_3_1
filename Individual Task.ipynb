В партии 10 % нестандартных деталей. Наудачу отобраны четыре детали. Написать биномиальный закон распределения дискретной случайной величины Х – числа нестандартных деталей среди четырёх отобранных и построить многоугольник распределения.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

pb = binom(n = 4, p = 0.1)

x = np.arange(0,5)
pmf = pb.pmf(x)
a = [round(i, 4) for i in pmf]
print(x)
print(a)
plt.plot(x , pb.pmf(x))
plt.ylabel('P(x)')
plt.xlabel('x')
plt.show()
```
